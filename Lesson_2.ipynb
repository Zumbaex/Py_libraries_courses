### Задание 1

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InLineBackend.figure_format = 'svg'
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [3.5, 3.8, 4.2, 4.5, 5, 5.5, 7]

fig, ax = plt.subplots(nrows = 1, ncols = 2)
ax1, ax2 = ax.flatten()
fig.set_size_inches(15, 6)
fig.subplots_adjust(wspace = 0.3)

ax1.plot(x, y)
ax2.scatter(x, y)
plt.show()
```

### Задание 2

```python
t = np.linspace(0, 10, 51)
f = np.cos(t)

fig, ax = plt.subplots()
fig.set_size_inches(8, 6)
plt.title('График f(t)', fontsize = 20)
plt.xlabel('Значения t', fontsize = 16)
plt.ylabel('Значения f', fontsize = 16)
plt.xlim([0.5, 9.5])
plt.ylim([-2.5, 2.5])

plt.plot(t, f, color = 'green')
plt.show()
```

### Задание 3

```python
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)

fig, ax = plt.subplots(nrows = 2, ncols = 2)
ax1, ax2, ax3, ax4 = ax.flatten()
fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace = 0.3, hspace = 0.3)

ax1.plot(x, y1)
ax1.set_title('График y1', fontsize = 16)
ax1.set_xlim(-5, 5)

ax2.plot(x, y2, color = 'yellow')
ax2.set_title('График y1', fontsize = 16)

ax3.plot(x, y3, color = 'red')
ax3.set_title('График y1', fontsize = 16)

ax4.plot(x, y4, color = 'green')
ax4.set_title('График y1', fontsize = 16)
```
